````python
Ø§Ù‚ØªØ±Ø§Ø­Ùƒ ØµØ­ÙŠØ­ 100%ØŒ ÙˆÙ‡Ø°Ø§ Ù‡Ùˆ **Ø§Ù„ØªØµØ±Ù Ø§Ù„Ø§Ø­ØªØ±Ø§ÙÙŠ** Ù‚Ø¨Ù„ ØªÙ‚Ø±ÙŠØ± Ø£ÙŠ Indexes.
Ø³Ø£Ø¹Ø·ÙŠÙƒ **ÙƒÙˆØ¯ ÙˆØ§Ø­Ø¯ ÙˆØ§Ø¶Ø­** ØªØ´ØºÙ‘Ù„Ù‡ ÙÙŠ Supabase (SQL Editor) ÙˆÙŠØ·Ù„Ø¹ Ù„Ùƒ:

* Ø§Ù„Ø¬Ø¯Ø§ÙˆÙ„
* Ø§Ù„Ø£Ø¹Ù…Ø¯Ø©
* Ø§Ù„Ù…ÙØ§ØªÙŠØ­ Ø§Ù„Ø£Ø³Ø§Ø³ÙŠØ©
* Ø§Ù„Ù…ÙØ§ØªÙŠØ­ Ø§Ù„Ø£Ø¬Ù†Ø¨ÙŠØ© (Ø§Ù„Ø¹Ù„Ø§Ù‚Ø§Øª)
* Ø§Ù„Ø¬Ø¯Ø§ÙˆÙ„ Ø§Ù„Ù…Ø±ØªØ¨Ø·Ø© Ø¨ÙƒÙ„ Ø¬Ø¯ÙˆÙ„

ÙˆØ¨Ø¹Ø¯Ù‡Ø§ ØªØ±Ø¬Ø¹ Ù„ÙŠ Ø§Ù„Ù†ØªÙŠØ¬Ø© ÙˆØ£Ù†Ø§ Ø£Ø¨Ù†ÙŠ Ù„Ùƒ **Ø£ÙØ¶Ù„ Index Strategy** Ù…Ø®ØµÙ‘ØµØ© Ù„Ø³ÙƒÙŠÙ…Ø§Ùƒ (Ù…Ø´ Ø¹Ø§Ù…Ø©).

---

## âœ… Ø§Ù„ÙƒÙˆØ¯ Ø§Ù„Ù†Ù‡Ø§Ø¦ÙŠ Ù„Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ù„Ø³ÙƒÙŠÙ…Ø§ + Ø§Ù„Ø¹Ù„Ø§Ù‚Ø§Øª (PostgreSQL / Supabase)

### 1ï¸âƒ£ Ø§Ù„Ø¬Ø¯Ø§ÙˆÙ„ + Ø§Ù„Ø£Ø¹Ù…Ø¯Ø© + Ù†ÙˆØ¹ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª

```sql
SELECT
    table_name,
    column_name,
    data_type,
    is_nullable
FROM information_schema.columns
WHERE table_schema = 'public'
ORDER BY table_name, ordinal_position;
```

---

### 2ï¸âƒ£ Ø§Ù„Ù…ÙØ§ØªÙŠØ­ Ø§Ù„Ø£Ø³Ø§Ø³ÙŠØ© (Primary Keys)

```sql
SELECT
    tc.table_name,
    kc.column_name
FROM information_schema.table_constraints tc
JOIN information_schema.key_column_usage kc
  ON tc.constraint_name = kc.constraint_name
WHERE tc.constraint_type = 'PRIMARY KEY'
  AND tc.table_schema = 'public'
ORDER BY tc.table_name;
```

---

### 3ï¸âƒ£ Ø§Ù„Ø¹Ù„Ø§Ù‚Ø§Øª Ø¨ÙŠÙ† Ø§Ù„Ø¬Ø¯Ø§ÙˆÙ„ (Foreign Keys) ğŸ”¥ Ø§Ù„Ø£Ù‡Ù…

```sql
SELECT
    tc.table_name        AS source_table,
    kcu.column_name      AS source_column,
    ccu.table_name       AS target_table,
    ccu.column_name      AS target_column
FROM information_schema.table_constraints tc
JOIN information_schema.key_column_usage kcu
  ON tc.constraint_name = kcu.constraint_name
JOIN information_schema.constraint_column_usage ccu
  ON ccu.constraint_name = tc.constraint_name
WHERE tc.constraint_type = 'FOREIGN KEY'
  AND tc.table_schema = 'public'
ORDER BY source_table;
```

---

### 4ï¸âƒ£ BONUS (Ø§Ø®ØªÙŠØ§Ø±ÙŠ): ÙƒÙ„ Ø§Ù„Ù€ Indexes Ø§Ù„Ø­Ø§Ù„ÙŠØ©

```sql
SELECT
    tablename,
    indexname,
    indexdef
FROM pg_indexes
WHERE schemaname = 'public'
ORDER BY tablename;
```

---

## ğŸ“Œ Ù…Ø§Ø°Ø§ ØªØ±Ø¬Ø¹ Ù„ÙŠØŸ

Ø§Ù†Ø³Ø® **Ù†ØªØ§Ø¦Ø¬ Ø§Ù„Ø§Ø³ØªØ¹Ù„Ø§Ù…Ø§Øª 3 Ùˆ 4** (Ø§Ù„Ø¹Ù„Ø§Ù‚Ø§Øª + Ø§Ù„Ù€ Indexes Ø§Ù„Ø­Ø§Ù„ÙŠØ©)
ÙˆÙ„Ùˆ Ø­Ø§Ø¨ØŒ Ø­ØªÙ‰ 1 ÙŠÙƒÙÙŠÙ†ÙŠ Ù„Ùˆ ÙƒØ§Ù†Øª Ø§Ù„Ø³ÙƒÙŠÙ…Ø§ Ù…Ø¹Ù‚Ø¯Ø©.

---

## Ù…Ø§Ø°Ø§ Ø³Ø£ÙØ¹Ù„ Ø¨Ø¹Ø¯Ù‡Ø§ØŸ

Ø¨Ù†Ø§Ø¡Ù‹ Ø¹Ù„Ù‰ Ø§Ù„Ù†ØªÙŠØ¬Ø© Ø³Ø£Ø¹Ø·ÙŠÙƒ:

* âœ… Indexes Ø¯Ù‚ÙŠÙ‚Ø© (Single / Composite)
* âœ… ØªØ±ØªÙŠØ¨ Ø§Ù„Ø£Ø¹Ù…Ø¯Ø© Ø¯Ø§Ø®Ù„ Ø§Ù„Ù€ Composite Index (Ù…Ù‡Ù… Ø¬Ø¯Ù‹Ø§)
* âœ… Indexes ÙŠØ¬Ø¨ Ø­Ø°ÙÙ‡Ø§ (Ø¥Ù† ÙˆØ¬Ø¯Øª ØªÙƒØ±Ø§Ø±Ø§Øª)
* âœ… Ù‡Ù„ ØªØ­ØªØ§Ø¬ Partial Index (`WHERE status = 'active'`)
* âœ… Ù‡Ù„ Ø§Ù„Ø£ÙØ¶Ù„ RPC Ø¨Ø¯Ù„ JOIN Ù„Ø¨Ø¹Ø¶ Ø§Ù„Ø­Ø§Ù„Ø§Øª

ÙƒÙ„ Ù‡Ø°Ø§ **Ù…Ø®ØµØµ Ù„Ø³ÙƒÙŠÙ…Ø§Ùƒ ÙÙ‚Ø·**ØŒ ÙˆÙ„ÙŠØ³ ÙƒÙ„Ø§Ù… Ø¹Ø§Ù….

Ù†ÙÙ‘Ø° Ø§Ù„ÙƒÙˆØ¯ØŒ ÙˆØ£Ø¹Ø¯ Ù„ÙŠ Ø§Ù„Ù†ØªØ§Ø¦Ø¬.
````
